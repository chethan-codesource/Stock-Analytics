```python
#This code is used to download data from the URL with increment date and save the csv file into the local machine:
#Libraries:

import datetime
from urllib import request
import requests
import pandas as pd
from dateutil.relativedelta import relativedelta

#Define function to download the stockdata from the URL and save it to our local machine:

def download_stock_data(csv_url,filename):
    try:        
        response = request.urlopen(csv_url)
        csv = response.read()
        csv_str = str(csv)
        lines = csv_str.split("\\n")
        dest_url = (r'C:\\Users\\Chethan\\Documents\\nifty_data\\')+(filename)#Mention the path file as to be saved with an increment filename:
        fx = open(dest_url,"w")
        for line in lines:
            fx.write(line + "\n")
#         fx.close()
    except Exception as e: 
        print("No data available")


#Here we mention date range Start and End date:

start_date = datetime.date(2015,1,1)
# edate = datetime.date(2021,3,5)
end_date = datetime.date.today()
delta = end_date - start_date
ran = (delta.days + 1)
for i in range(ran):
    date1 = start_date
    date1 += datetime.timedelta(days=i)
    validate = date1.strftime("%d%m%Y")
    filename = 'ind_close_all_' + (validate)+'.csv'
    day_url = 'https://www1.nseindia.com/content/indices/ind_close_all_' + (validate)+'.csv' #URl used to download data with increment date:
    download_stock_data(day_url,filename)#Calling function to download data:

#Print URL list with increment date and also print if data is not available ("data not available" on weekends and government holidays)
    print(day_url)
```
